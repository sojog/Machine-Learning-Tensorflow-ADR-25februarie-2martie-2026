```python
df["medv"] # iau o coloana ca serie
```

```python
y = df[["medv"]] # iau o coloana ca DataFrame
y
```

```python
x = df[['indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax', 'ptratio',
       'lstat']]
x
```

```python
x = df.drop("medv", axis=1)
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test =  train_test_split(x, y, test_size=0.2)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
```

```python
linear_model = LinearRegression()
linear_model
```

```python
knn_model = KNeighborsRegressor()
knn_model
```

```python
linear_model.fit(x_train, y_train)
```

```python
knn_model.fit(x_train, y_train)
```

```python
linear_model.score(x_test, y_test)
```

```python
knn_model.score(x_test, y_test)
```

```python
scores = {}
for k in range(2, 11): 
    inner_knn_model = KNeighborsRegressor(n_neighbors=k)
    inner_knn_model.fit(x_train, y_train)
    inner_score = inner_knn_model.score(x_test, y_test)
    scores[k] = inner_score

scores
```
