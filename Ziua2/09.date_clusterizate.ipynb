```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("categories_datasets/2_categories.csv")
df
```

```python
unsupervised_df = df.drop("category", axis=1)
```

```python
plt.scatter(unsupervised_df["coord_x"], unsupervised_df["coord_y"] )
```

```python
plt.scatter(df["coord_x"], df["coord_y"] )
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans_model = KMeans(n_clusters = 2)
kmeans_model
```

```python
kmeans_model.fit(X=unsupervised_df)
```

```python
predicted_categories = kmeans_model.predict(unsupervised_df)
predicted_categories
```

```python
plt.scatter(unsupervised_df["coord_x"], unsupervised_df["coord_y"], c=predicted_categories )
```

```python
plt.scatter(df["coord_x"], df["coord_y"], c=df["category"].map({"a":"red", "b":"blue"}))
```

```python
df_4cat = pd.read_csv("categories_datasets/4_categories.csv")
df_4cat
```

```python
kmeans_model_4_cat = KMeans(n_clusters = 4)
kmeans_model_4_cat
```

```python
kmeans_model_4_cat.fit(X=df_4cat[["coord_x", "coord_y"]])
```

```python
predicted_categories_results = kmeans_model_4_cat.predict(df_4cat[["coord_x", "coord_y"]])
```

```python
plt.scatter(df_4cat["coord_x"], df_4cat["coord_y"], c=predicted_categories_results)
```

```python
plt.scatter(df_4cat["coord_x"], df_4cat["coord_y"], c=df_4cat["category"].map({"a":"red", "b":"blue", "c":"green", "d":"pink"}))
```
