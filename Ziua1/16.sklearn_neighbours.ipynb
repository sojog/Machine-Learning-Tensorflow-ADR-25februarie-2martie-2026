```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# mp, suprafata
apartamente = np.array([
    (40, 65_000), (49, 71_000), (50, 75_000), (57, 83_000), (60, 120_000), (73, 154_000), (85, 170_000), (91, 210_000), (100, 300_000)
])
```

```python
suprafete = apartamente[:, 0]
preturi = apartamente[:, 1]
```

```python
plt.scatter(suprafete, preturi, c="red")
plt.plot(suprafete, preturi);
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_model = LinearRegression()
linear_model
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn_model = KNeighborsRegressor()
knn_model
```

```python
suprafete_2d = suprafete.reshape(-1, 1)
```

```python
linear_model.fit(suprafete_2d, preturi)
```

```python
knn_model.fit(suprafete_2d, preturi)
```

```python
plt.scatter(suprafete, preturi , c="red")
plt.plot(suprafete, linear_model.coef_ * suprafete + linear_model.intercept_, c="black");
# plt.plot(suprafete, knn_model.coef_ * suprafete + knn_model.intercept_, c="black");
```

```python
suprafata = 67
```

```python
linear_model.predict([[suprafata]])
```

```python
knn_model.predict([[suprafata]])
```

```python
plt.scatter(suprafete, preturi , c="red")
plt.plot(suprafete, linear_model.coef_ * suprafete + linear_model.intercept_, c="black");

plt.scatter(suprafata, linear_model.predict([[suprafata]]), c="purple")
plt.scatter(suprafata, knn_model.predict([[suprafata]]), c="green");
```

```python
plt.scatter(suprafete, preturi , c="red")
plt.plot(suprafete, linear_model.coef_ * suprafete + linear_model.intercept_, c="black");

suprafata = 80
plt.scatter(suprafata, linear_model.predict([[suprafata]]), c="purple")
plt.scatter(suprafata, knn_model.predict([[suprafata]]), c="green");
```
