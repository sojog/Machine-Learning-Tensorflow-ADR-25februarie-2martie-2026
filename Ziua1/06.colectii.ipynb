# Iteratia se face printr-o colectie

# 1. Range -> interval de la un numar la altul

```python
for i in range(10):
    print("i=", i)
```

```python
for i in range(6, 10):
    print("i=", i)
```

```python
for i in range(6, 10, 2):
    print("i=", i, type(i))
```

# 2. Lista -> colectie heterogena, mutabila

```python
lista = [20, 10, 4321.321, "hello", True, ["31", 30]]

for i in lista:
    print("i=", i, type(i))
```
