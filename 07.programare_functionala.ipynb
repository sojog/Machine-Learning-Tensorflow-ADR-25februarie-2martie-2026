# 1. Map

```python
o_lista = [23, 32, 15, 51, 20, 30]
```

```python
def dubleaza(x):
    return x * 2
```

```python
map(dubleaza, o_lista)
```

```python
list(map(dubleaza, o_lista))
```

# List comprehention - specific Python

```python
o_lista = [23, 32, 15, 51, 20, 30]

o_noua_lista = [i * 2 for i in o_lista ]
o_noua_lista
```

# 2. Filter

```python
o_lista = [23, 32, 15, 51, 20, 30]

def este_mai_mare_ca_20(x):
    return x > 20
```

```python
list(filter(este_mai_mare_ca_20, o_lista))
```

```python
o_lista = [23, 32, 15, 51, 20, 30]

list(filter(lambda x: x>20, o_lista))
```

# List comprehention - filtrare specifica Python

```python
o_lista = [23, 32, 15, 51, 20, 30]

o_noua_lista = [i  for i in o_lista if i > 20 ]
o_noua_lista
```

```python
o_lista = [23, 32, 15, 51, 20, 30]

o_noua_lista = [i if i > 20 else "__"  for i in o_lista ]
o_noua_lista
```

# 3. Reduce

```python
o_lista = [23, 32, 15, 51, 20, 30]
sum(o_lista), min(o_lista), max(o_lista), len(o_lista)
```

```python
def functie_reducere(x, y):
    print("x=", x, "y=", y)
    return x if x > y else y 
```

```python
from functools import reduce
```

```python
reduce(functie_reducere, o_lista)
```

```python
reduce(lambda x, y: x if x > y else y, o_lista)
```

```python
max(o_lista)
```

```python
reduce(lambda x, y: x + y, o_lista)
```
