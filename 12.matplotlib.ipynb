```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.scatter(10, 15)
plt.show()
```

```python
plt.scatter([10, 15], [30, 35]);
```

```python
plt.scatter([10, 15], [30, 35]);
plt.plot([10, 15], [30, 35]);
```

```python
plt.scatter([10, 15, 23], [30, 35, 40]);
plt.plot([10, 15, 23], [30, 35, 40]);
```

```python
x = np.array([10, 15, 23])
y = np.array([30, 35, 40])

plt.scatter(x, y);
plt.plot(x, y);
```

```python
x = np.array([10, 15, 23])
y = np.array([30, 35, 40])

plt.bar(x, y)
```

```python
x = np.array([10, 15, 23])
y = np.array([30, 35, 40])

plt.barh(x, y);
```

```python
x = np.array([10, 15, 25])

plt.pie(x, labels=["albastru",  "portocaliu", "verde"]);
```
