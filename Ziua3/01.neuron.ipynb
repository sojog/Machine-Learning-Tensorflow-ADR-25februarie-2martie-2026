# Cum facem cu OR clasic?

```python
def OR_clasic(x1, x2):
    print(x1, "OR", x2, "=", x1 or x2)
```

```python
OR_clasic(True, True)
OR_clasic(False, True)
OR_clasic(True, False)
OR_clasic(False, False)
```

```python
OR_clasic(1, 1)
OR_clasic(0, 1)
OR_clasic(1, 0)
OR_clasic(0, 0)
```

```python
def functie_activare(x):
    return 0 if x <= 0 else 1 

def OR_neuron(x1, x2):
    w1 = 1
    w2 = 1 
    b = 0 # A treia schimbare?? 
    rezultat_calcul =  x1 * w1 + x2 * w2 + b
    rezultat_dupa_functie_de_activare =  functie_activare(rezultat_calcul)
    print(x1, "OR", x2, "=", rezultat_dupa_functie_de_activare)

OR_neuron(1, 1)
OR_neuron(0, 1)
OR_neuron(1, 0)
OR_neuron(0, 0)
```

# Este posibil sa avem acelasi rezultat cu alte valori?

```python
def functie_activare(x):
    return 0 if x < 0 else 1  # Am schimbat functia de activare (de la x <=0 la x < 0)

def OR_neuron(x1, x2):
    w1 = 1
    w2 = 1 
    b = -2 # Am ajustat aici ponderea pentru a avea rezulate similare
    rezultat_calcul =  x1 * w1 + x2 * w2 + b
    rezultat_dupa_functie_de_activare =  functie_activare(rezultat_calcul)
    print(x1, "OR", x2, "=", rezultat_dupa_functie_de_activare)

OR_neuron(1, 1)
OR_neuron(0, 1)
OR_neuron(1, 0)
OR_neuron(0, 0)
```
