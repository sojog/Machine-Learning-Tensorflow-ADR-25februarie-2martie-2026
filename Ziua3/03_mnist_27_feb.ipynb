```python
((x_train, y_train ), (x_test, y_test)) = mnist_dataset
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
x_train_scaled = x_train / 255
plt.imshow(x_train_scaled[0])
```

```python
x_train_scaled.max()
```

```python
x_train_scaled[0].shape
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(784,))
])
```

```python
model = tf.keras.Sequential([
    ## INPUT LAYER -  va fi intotdeauna la fel in functie de tipul datelor de intrare
    tf.keras.layers.Flatten(input_shape=(28,28)),

    tf.keras.layers.Dense(100, activation="relu"),


    ## OUTPUT Layer - va fi intotdeauna la fel in functie de cum trebuie sa arate rezultatul
    tf.keras.layers.Dense(10, activation="softmax")

])
```

```python
model.compile(loss="sparse_categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
model_history = model.fit(x_train_scaled, y_train, epochs=10)
model_history
```

```python
x_test_scaled = x_test / 255
```

```python
model.evaluate( x_test_scaled, y_test)
```

```python
plt.imshow(x_test_scaled[0])
```

```python
y_test_predicted = model.predict(x_test_scaled)
y_test_predicted
```

```python
loss = model_history.history['loss']
plt.plot(loss)

plt.title("Loss ")


plt.xlabel("10 epoci")
plt.ylabel("Loss")
plt.legend(["Training Data", "Validation Data"])
```
