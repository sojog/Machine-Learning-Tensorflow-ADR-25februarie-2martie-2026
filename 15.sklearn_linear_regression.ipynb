```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# mp, suprafata
apartamente = np.array([
    (40, 65_000), (49, 71_000), (50, 75_000), (57, 83_000), (60, 120_000), (73, 154_000), (85, 170_000), (91, 210_000), (100, 300_000)
])
```

```python
suprafete = apartamente[:, 0]
preturi = apartamente[:, 1]
```

```python
plt.scatter(suprafete, preturi, c="red")
plt.plot(suprafete, preturi);
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model
```

```python
suprafete
```

```python
suprafete_2d = suprafete.reshape(-1, 1)
suprafete_2d
```

```python
model.fit(suprafete_2d, preturi)
```

```python
model.coef_, model.intercept_
```

```python
# comparam cu valorile din scipy
slope=np.float64(3603.603034134008)
intercept=np.float64(-103575.5372945639)
```

```python
plt.scatter(suprafete, preturi , c="red")
plt.plot(suprafete, model.coef_ * suprafete + model.intercept_, c="black");
```
