```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
cat_image = cv2.imread("cat.jpeg")
```

```python
type(cat_image)
```

```python
plt.imshow(cat_image)
```

```python
gray_scale_cat_image = cv2.imread("cat.jpeg", cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(gray_scale_cat_image, cmap="gray");
```

```python
gray_scale_cat_image.shape
```

```python
resized_gray_scale_cat_image = cv2.resize(gray_scale_cat_image,  (192 // 3, 204 // 3))
resized_gray_scale_cat_image.shape
```

```python
plt.imshow(resized_gray_scale_cat_image, cmap="gray");
```

```python
blurred_image = cv2.blur(gray_scale_cat_image, (3, 3))
cv2.imwrite("blurred_cat2.png", blurred_image)
```

```python
blurred_image = cv2.blur(gray_scale_cat_image, (7,7))
cv2.imwrite("blurred_cat3.png", blurred_image)
```

```python
left_sobel_kernel = np.array([[1, 0, -1], [2, 0, -2], [1, 0, -1]])
left_sobel_kernel
```

```python
left_kernel_image = cv2.filter2D(gray_scale_cat_image, -1, left_sobel_kernel)
plt.imshow(left_kernel_image, cmap="gray")
```

```python
left_sobel_kernel.T
```

```python
top_sobel_kernel = left_sobel_kernel.T
```

```python
top_kernel_image = cv2.filter2D(gray_scale_cat_image, -1, top_sobel_kernel)
plt.imshow(top_kernel_image, cmap="gray")
```
