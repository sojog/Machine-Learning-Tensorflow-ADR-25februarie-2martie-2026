```python
architectura_le_net5 = tf.keras.models.Sequential(
    [
        # Input shape depinde de cum arata imaginea
        tf.keras.layers.Input(shape=(28, 28, 1)),

        ## Rescalez imaginea de la 0..255 la maxim 0..1
        tf.keras.layers.Rescaling(scale=1./255),

        ## S-a inceput cu tanh, acum relu este varianta mai buna
        tf.keras.layers.Conv2D(kernel_size=(5, 5), filters=6, activation="tanh"),

        ## Cea mai buna practica este MaxPooling2D
        tf.keras.layers.AveragePooling2D(pool_size=(2, 2), strides=(2, 2)),

          ## S-a inceput cu tanh, acum relu este varianta mai buna
        tf.keras.layers.Conv2D(kernel_size=(5, 5), filters=6, activation="tanh"),

        ## Cea mai buna practica este MaxPooling2D
        tf.keras.layers.AveragePooling2D(pool_size=(2, 2), strides=(2, 2)),


        # Architectura de fully connected
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(128, activation='tanh'),
        tf.keras.layers.Dense(10, activation='softmax')


    ]
)

architectura_le_net5.compile(optimizer="sgd", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
architectura_le_net5.summary()
```

```python
imbunatatire_le_net5 = tf.keras.models.Sequential(
    [
        # Input shape depinde de cum arata imaginea
        tf.keras.layers.Input(shape=(28, 28, 1)),

        ## Rescalez imaginea de la 0..255 la maxim 0..1
        tf.keras.layers.Rescaling(scale=1./255),

        ## S-a inceput cu tanh, acum relu este varianta mai buna
        tf.keras.layers.Conv2D(kernel_size=(3, 3), filters=6, activation="relu"),

        ## Cea mai buna practica este MaxPooling2D
        tf.keras.layers.MaxPooling2D(pool_size=(2, 2)),

          ## S-a inceput cu tanh, acum relu este varianta mai buna
        tf.keras.layers.Conv2D(kernel_size=(3, 3), filters=6, activation="relu"),

        ## Cea mai buna practica este MaxPooling2D
        tf.keras.layers.MaxPooling2D(pool_size=(2, 2)),


        # Architectura de fully connected
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(128, activation='relu'),
        tf.keras.layers.Dense(10, activation='softmax')


    ]
)

imbunatatire_le_net5.compile(optimizer="sgd", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
imbunatatire_le_net5.summary()
```

```python
EPOCHS = 15
model_history_original = architectura_le_net5.fit(x_train, y_train, validation_split=0.2, epochs=EPOCHS, verbose=0)
model_history_upgraded = imbunatatire_le_net5.fit(x_train, y_train, validation_split=0.2, epochs=EPOCHS, verbose=0)
```

```python
from re import L
history_dict = model_history_original.history

loss = history_dict['loss']
val_loss = history_dict['val_loss']

epochs = range(1, len(loss) + 1)

plt.plot(epochs, loss, "bo", label="Training loss")
plt.plot(epochs, val_loss, "b", label="Validation loss")
plt.legend()
plt.title("Training and validation loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.show()
```

```python
history_dict = model_history_upgraded.history

loss = history_dict['loss']
val_loss = history_dict['val_loss']

epochs = range(1, len(loss) + 1)

plt.plot(epochs, loss, "bo", label="Training loss")
plt.plot(epochs, val_loss, "b", label="Validation loss")
plt.legend()
plt.title("Training and validation loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.show()
```

```python
architectura_le_net5.evaluate(x_test, y_test)
```
